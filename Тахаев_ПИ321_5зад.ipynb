```python
#Импорт библиотеки для работы с таблицами
import pandas as pd

#Читаем нашу таблицу в перменную rosn, где копируем путь к файлу для использования, как аргумент функции read_cvs 
rosn = pd.read_csv("/content/ROSN Historical Data.csv", parse_dates = ['Date'], usecols = range(5), index_col = 0)

#Проверяем правильность чтения данных
rosn.head(3)
```

```python
#Тип данных по столбцам
rosn.dtypes
```

```python
#Переименуем столбец с именем Price на Close
rosn.rename(columns = {"Price" : "Close"}, inplace = True)

#Проверим, выполнились ли изменения
rosn.head(3)
```

```python
#Импорт библиотеки для построения биржевых графиков
import mplfinance as mpf

#Импорт библиотеки для исключения ошибки plt.show()
import matplotlib.pyplot as plt
```

```python
#Построение свечного графика
#Нанесение скользящих средних с периодами 51 и 200 дней
mpf.plot(rosn, warn_too_much_data = 10000, figsize = (18, 6), mav = (51, 200), title = "ROSN + SMA 51 + SMA 200", type = "candle")
plt.show()
```
